```
import fast_bert
databunch = fast_bert.data_cls.BertDataBunch( data_dir=args['data_dir'], 
                        label_dir=LABEL_PATH,
                        tokenizer=args.model_name, 
                        train_file='train.csv', 
                        val_file='valid.csv',
                        test_data='test.csv', 
                        label_file="labels.csv",
                        text_col="comment_text", 
                        label_col=label_cols,
                        batch_size_per_gpu=args['train_batch_size'],
                        max_seq_length=args['max_seq_length'],
                        multi_gpu=args.multi_gpu,
                        multi_label=True,
                        model_type=args.model_type)
```

```
metrics = [{'name': 'accuracy', 'function': accuracy_multilabel}]
```

```
metrics = []
metrics.append({'name': 'accuracy_thresh', 'function': accuracy_thresh})
metrics.append({'name': 'roc_auc', 'function': roc_auc})
metrics.append({'name': 'fbeta', 'function': fbeta})
```

```
learner = BertLearner.from_pretrained_model(dataBunch = databunch, 
                       pretrained_path = args.model_name,
                       metrics = metrics, 
                       device = device, 
                       logger = logger,
                       output_dir = args.output_dir,
                       finetuned_wgts_path = FINETUNED_PATH,
                       warmup_steps = args.warmup_steps, 
                       is_fp16=args.fp16, 
                       loss_scale=args.loss_scale,
                       logging_steps = 0, 
                       multi_gpu=args.multi_gpu,  
                       multi_label=True)
```

```
learner.fit(args.num_train_epochs,lr=args.learning_rate, schedule_type="warmup_linear",validate = True)
```

```
learner.validate()
```
